```python
from collections import defaultdict
import xml.etree.ElementTree as ET
import xmltodict
import json
import os

brand_names_count = defaultdict(int)

for i in os.walk('/home/sumit-tyagi/sumit-ssd/sumit/data/annotations'):
    print(i[2])
    for j in i[2]:
        tree = ET.parse('/home/sumit-tyagi/sumit-ssd/sumit/data/annotations/'+j)
        xml_data = tree.getroot()
        xmlstr = ET.tostring(xml_data, encoding='utf-8', method='xml')

        data_dict = dict(xmltodict.parse(xmlstr))
        for i in data_dict['annotation']['object']:
            brand_names_count[i['name']] += 1
            i['name'] = 'packet'


```

```python
import pandas as pd
brand_names_list = list(brand_names_count.keys())
annotation_data = pd.DataFrame(data = list(brand_names_count.items()), columns= ['brand_name', 'annotation_count'])
```

```python
df = pd.read_csv('/home/sumit-tyagi/sumit-ssd/sumit/data/csvs/products.csv')
df.columns = ['id', 'brand_name', 'count', "#units_charged", 'loaded_weights']
df.head(2)
```

```python
def fun(x):
    x = x.lower()
    x = x.replace(" ", "_")
    x = x.replace("/", "_")
    return x
    
df.brand_name = df.brand_name.apply(fun)
```

```python
total_data = annotation_data.merge(df[['brand_name', 'count']], on='brand_name', how='outer' )
```

```python
total_data[total_data['annotation_count'].isna()].sort_values(by='brand_name')
```

```python
total_data[total_data['count'].isna()].sort_values(by='brand_name')
```

```python
data_processed = annotation_data.merge(df[['brand_name', 'count']], on='brand_name', how='inner' )
data_processed.annotation_count =(data_processed.annotation_count - data_processed.annotation_count.min())/(data_processed.annotation_count.max() - data_processed.annotation_count.min())
data_processed['count'] =(data_processed['count'] - data_processed['count'].min())/(data_processed['count'].max() - data_processed['count'].min())
```

```python
import plotly.graph_objects as go

temp = data_processed.sort_values(by='count', ascending = False)
x = temp.brand_name
fig = go.Figure()
fig.add_trace(go.Bar(
    x=x,
    y=temp['count'],
    name='Count in Pepsi Database',
    marker_color='indianred'
))
fig.add_trace(go.Bar(
    x=x,
    y=temp.annotation_count,
    name='Count in Annotated Images',
    marker_color='lightsalmon'
))

# Here we modify the tickangle of the xaxis, resulting in rotated labels.

fig.update_layout(barmode='group', xaxis_tickangle=-45, title={'text': 'Normalised Product Count'})
fig.show()

fig.write_html('/home/sumit-tyagi/sumit-ssd/sumit/eda/story-telling/normalised_product_count_annotation_vs_csv_bar_plot.html')
```

```python
import cv2

aspect_ratios = defaultdict(int)

for i in os.walk('/home/sumit-tyagi/sumit-ssd/sumit/data/client-data-images'):
    #print(i[2])
    for j in i[2]:
        img = cv2.cv2.imread('/home/sumit-tyagi/sumit-ssd/sumit/data/client-data-images/'+j)
        aspect_ratios[j] = float(img.shape[1]) / img.shape[0] # width/ height
            

import plotly.figure_factory as ff


hist_data = [list(aspect_ratios.values())]
group_labels = ['Aspect Ratio'] # name of the dataset

fig = ff.create_distplot(hist_data, group_labels)
fig.update_layout(title={'text': 'Distribution of Aspect Ratio of Images  (width/ height)'})
fig.write_html('/home/sumit-tyagi/sumit-ssd/sumit/eda/story-telling/distribution_of_aspect_ratio_of_images.html')
fig.show()
```

```python
from collections import defaultdict
import xml.etree.ElementTree as ET
import xmltodict
import json
import os

annotation_aspect_ratios = []

for i in os.walk('/home/sumit-tyagi/sumit-ssd/sumit/data/annotations'):
    #print(i[2])
    for j in i[2]:
        tree = ET.parse('/home/sumit-tyagi/sumit-ssd/sumit/data/annotations/'+j)
        xml_data = tree.getroot()
        xmlstr = ET.tostring(xml_data, encoding='utf-8', method='xml')

        data_dict = dict(xmltodict.parse(xmlstr))
        for i in data_dict['annotation']['object']:
            annotation_aspect_ratios.append((float(i['bndbox']['xmax']) - float(i['bndbox']['xmin']))/ ((float(i['bndbox']['ymax']) - float(i['bndbox']['ymin']))))


import plotly.figure_factory as ff


hist_data = [annotation_aspect_ratios]
group_labels = ['Annotation Aspect Ratio'] # name of the dataset

fig = ff.create_distplot(hist_data, group_labels)
fig.update_layout(title={'text': 'Distribution of Aspect Ratios of Annotations in Images  (width/ height)'})
fig.write_html('/home/sumit-tyagi/sumit-ssd/sumit/eda/story-telling/distribution_of__annotations_aspect_ratio_of_images.html')
fig.show()
```
